```python
import os
import sys
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as signal
from scipy.misc import electrocardiogram
from scipy.signal import find_peaks
```

```python
file = open("data.txt", "r")
y = []
x = []
dict = {}
i = 0
```

```python
for line in file:
    if i > 5: # Disregard header info
        index = i-6
        lineParts = line.strip().split(",") # Split the line into each columns values
        dict[index] = [lineParts[1],lineParts[2],lineParts[3],lineParts[4]] # Create a dictionary entry w/ the sensor values

    i += 1
```

```python
x = list(dict.keys())
y = []
for key,value in dict.items(): # Create an array with the dictionary values
    y.append(abs(float((value[1]))))
plt.scatter(x, y)
plt.show()
```

```python
y = np.asarray(y)
thresh = 2000
peaks, _ = find_peaks(y, height=thresh, distance=100)
plt.plot(y)
plt.plot(peaks, y[peaks], "x")
plt.plot(np.full((len(y),1), thresh), "--", color="gray")
plt.title("Number of tics: " + str(len(peaks)))
plt.show()
```
